```python

def matrix_multiply(A, B):
    if len(A[0])!= len(B):
        print("Error, matrices are not compatible for multiplication.")
        return

    result = []
    for i in range(len(A)):
        row = []
        for j in range(len(B[0])):
            s = 0
            for k in range(len(A[0])):
                s = A[i][k] * B[k][j]
            row.append(s)
        result.append(row)

    print("Resultant matrix: ")
    for r in result:
        print(r)
                
        
```
